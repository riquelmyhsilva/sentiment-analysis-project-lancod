```python
import spacy
import pandas as pd

# Carregando o modelo de idioma em português
nlp = spacy.load('pt_core_news_sm')
df_class_words = pd.read_csv('sentilex_new.csv', sep = ';')
```

```python
# Texto de exemplo
text = """A inteligência artificial (IA) está revolucionando a maneira como interagimos com a tecnologia. 
O processamento de linguagem natural (PLN) é uma área crucial da IA, focada em permitir que as máquinas compreendam,
interpretem e gerem a linguagem humana. Tokenização é o primeiro passo nesse processo, onde uma sequência de caracteres
(como este parágrafo) é dividida em unidades menores, chamadas tokens. Estes tokens podem ser palavras, subpalavras,
ou até mesmo caracteres individuais, dependendo do algoritmo utilizado. Por exemplo,
a frase "IA está revolucionando" pode ser tokenizada em ["IA", "está", "revolucionando"]."""

# Aplicando a tokenização
doc = nlp(text)

# Extraindo os tokens
tokens = [token.text for token in doc]
df_text_tokens = pd.DataFrame(tokens)
df_text_tokens.columns = ['word']
```

```python
# Coletando informações dos tokens em um único DataFrame
dados_tokens = []
for token in doc:
    dados_tokens.append({
        'Token': token.text,
        'Índice de início': token.idx,
        'É pontuação?': token.is_punct,
        'É espaço em branco?': token.is_space,
        'Parte do discurso': token.pos_,
        'Dependência sintática': token.dep_
    })

# Criando DataFrame com os dados dos tokens
df_tokens = pd.DataFrame(dados_tokens)

# Exibindo o DataFrame
df_tokens
```

```python
import unicodedata

def norm(s):
    if not isinstance(s, str):
        return s
    s = s.lower()
    s = ''.join(c for c in unicodedata.normalize('NFD', s)
                if unicodedata.category(c) != 'Mn')
    return s
```

```python
df_class_words["word"] = df_class_words["word"].apply(norm)
```

```python
df_class_words.head()
```

```python
df_text_tokens["word"] = df_text_tokens["word"].apply(norm)
```

```python
df_text_tokens.head()
```

```python
merged = df_text_tokens.merge(
    df_class_words[['word', 'polarity_n0','polarity_n1']],
    on='word',
    how='left'
)
```
